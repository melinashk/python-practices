```python
# Assignment
def cal():
    a = int(input("Enter a = "))
    b = int(input("Entre b = "))
    o = input("Enter + - * /   ")

    if o == '+':
        print(a+b)
    elif o == '-':
        print(a-b)
    elif o == '*':
        print(a*b)

    elif (o == '/'):
        if b !=0:
            print(a/b)
        else:
            print("Value of b cant be zero")

    else:
        print("Invalid operator")
        
    x = input("Enter y for mor calculations = ")
    if x == 'y':
        cal()
        
cal()
```

```python
# Assignment
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b
def cal():
    a = int(input("Enter a = "))
    b = int(input("Entre b = "))
    o = input("Enter + - * /   ")

    if o == '+':
        print(add(a,b))
    elif o == '-':
        print(sub(a,b))
    elif o == '*':
        print(mul(a*b))

    elif (o == '/'):
        if b !=0:
            print(div(a,b))
        else:
            print("Value of b cant be zero")

    else:
        print("Invalid operator")
        
    x = input("Enter y for mor calculations = ")
    if x == 'y':
        cal()
        
cal()
```

```python
# and or  & | 
name = input("Enter name = ")
physics = int(input("Enter marks of physics="))
chemistry = int(input("Enter marks of chemistry="))
math = int(input("Enter marks of math="))
english = int(input("Enter marks of english="))
nepali = int(input("Enter marks of nepali="))
total = physics+chemistry+math+english+nepali
per = total/5
if per >= 80:
    print("The grade is A")
elif per >= 60:
    print("The grade is B")
elif per>=50:
    print("The grade is C")
elif per>=40:
    print("The grade is D")
else:
    print("You are failed")
```

```python
# and or  & | 
def result():
    name = input("Enter name = ")
    physics = int(input("Enter marks of physics="))
    chemistry = int(input("Enter marks of chemistry="))
    math = int(input("Enter marks of math="))
    english = int(input("Enter marks of english="))
    nepali = int(input("Enter marks of nepali="))
    total = physics+chemistry+math+english+nepali
    per = total/5
    if (per >= 80) | (per <=100):
        print("The grade is A")
    elif per >= 60:
        print("The grade is B")
    elif per>=50:
        print("The grade is C")
    elif per>=40:
        print("The grade is D")
    else:
        print("You are failed")
        
    x= input("Enter y for more result ")
    if x =='y':
        result()

        
result()
```
