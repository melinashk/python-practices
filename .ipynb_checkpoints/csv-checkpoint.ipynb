```python
infos = {}
total =0
names = []
prices = []
quantitys = []
total_prices = []
n = int(input("Enter how many types of product = "))
for i in range(1,n+1):
    name = input("Enter name=")
    price = int(input("Enter price"))
    quantity = int(input("Enter quantity"))
    
    
    total_price = price*quantity
    total = total + total_price
    
    names.append(name)
    prices.append(price)
    quantitys.append(quantity)
    total_prices.append(total_price)
    
infos = {'Name':names,'Price':prices,'Quantity':quantitys,'Total':total_prices}

print(infos)
import pandas as pd
df = pd.DataFrame(infos)
df.to_csv('billing.csv')
```

```python
import csv
n = int(input("Enter value of n = "))

with open('class11amraw.csv', 'a') as f:
    x = csv.writer(f)
    for i in range(1,n+1):
        name =input("Enter name = ")
        age = input("Enter age = ")
        add = input("Enter add = ")
        phone = input("Enter phone = ")
        fields = [i,name,age,add,phone]
        x.writerow(fields)
```
