```
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('http://www.nepalstock.com/floorsheet', 'html.parser')
response = requests.get(url)
response
```

```
data = response.text
data = BeautifulSoup(data, "lxml")
data
```

```
data.table
tag = data.find_all('tr')
```

```
tags = tag[1:-3]
```

```
output = []
for i in tags:
  output.append(i.text)

output
```

```
outputs = []
for i in output:
  outputs.append(i.split('\n')[1:-1])

outputs
```

```
import csv
n = len(outputs)

with open('data2.csv', 'a') as f:
    x = csv.writer(f)
    for i in range(n):
        x.writerow(outputs[i])
```
